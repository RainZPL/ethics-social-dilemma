```python
import numpy as np
import pandas as pd
from collections import defaultdict


def Utilitariansm(users):
    total_happiness = {}
    for user in users:
        for key,value in user['happiness'].items():
            if key in total_happiness:
                total_happiness[key] += value
            else:
                total_happiness[key] = value

    
    print(max(total_happiness,key=total_happiness.get))
    print(max(total_happiness.values()))


def Prioritarianism(users):
    min_happiness = defaultdict(lambda: float('inf'))
    total_happiness = 0
    for user in users:
        for key,value in user['happiness'].items():
            min_happiness[key] = min(min_happiness[key], value)
    
    # print(min_happiness)

    k = max(min_happiness,key=min_happiness.get)

    for user in users:
        for key,value in user['happiness'].items():
            if key == k:
                total_happiness += value       
    print(k)
    print(total_happiness)


def DesertBased_Proportionalism(users):
    weight_df = []
    for user in users:
        diff_user = []
        for value in user['contribution'].values():
            diff_user.append(value - user['luck'])
        weight_df.append(diff_user)
    
    print(weight_df)
    


def Libertarian_Proportionalism(users):
    max_contribution = defaultdict(lambda: float('-inf'))
    for user in users:
        for key,value in user['contribution'].items():
            max_contribution[key] = max(max_contribution[key], value)
    
    print(max_contribution)
    print(max(max_contribution,key=max_contribution.get))
    print(max(max_contribution.values()))



def envy_free(users):
    user_preferences = []
    for user in users:
        utility = []
        for value in user['happiness'].values():
            utility.append(value)
        user_preferences.append(utility)
        
    
    # 将用户偏好值转换为矩阵形式
    utilities = np.array(user_preferences)
    print(utilities)

    # 计算用户数量和物品数量
    num_users, num_items = utilities.shape
    """
    寻找envy-freeness分配方案
    
    :param utilities: 用户偏好矩阵
    :return: 分配方案，每个用户被分配的物品的索引
    """
    allocation = np.zeros(num_users, dtype=int)
    print(allocation)
    for i in range(num_users):
        # 找到当前用户的最佳分配
        max_utility = -np.inf
        max_item = -1
        for j in range(num_items):
            if j not in allocation or j == 0:
                user_utility = np.sum(utilities[i, :])
                print(utilities[i, :])
                print(user_utility)
                item_utility = utilities[i, j]
                print(item_utility)
                if item_utility > max_utility and item_utility <= user_utility:
                    max_utility = item_utility
                    max_item = j
        # 将最佳物品分配给当前用户
        allocation[i] = max_item
    return allocation



def DDE(good_action, bad_effect_intended, good_effect_intended, good_effect_weight, bad_effect_weight):
    if not good_action:
        return False
    if bad_effect_intended:
        return False
    if not good_effect_intended:
        return False
    if good_effect_weight <= bad_effect_weight:
        return False
    return True
```
