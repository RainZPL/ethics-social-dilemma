```python
import numpy as np
import pandas as pd
from collections import defaultdict
# from tabulate import tabulate


def Utilitarianism(users):
    total_happiness = {}
    for user in users:
        for key,value in user['happiness'].items():
            if key in total_happiness:
                total_happiness[key] += value
            else:
                total_happiness[key] = value

    tmp = []
    for val in total_happiness.values():
        tmp.append(val)
    
    if tmp.count(max(total_happiness.values()))>1 :
        print('This Principle CAN NOT give the BEST output. Please try another.')
    else:
        print(max(total_happiness,key=total_happiness.get))
        print(max(total_happiness.values()))


def Prioritarianism(users):
    min_happiness = defaultdict(lambda: float('inf'))
    for user in users:
        for key,value in user['happiness'].items():
            min_happiness[key] = min(min_happiness[key], value)
    
    print(min_happiness)
    tmp = []
    for val in min_happiness.values():
        tmp.append(val)
    
    if tmp.count(max(min_happiness.values()))>1 :
        print('This Principle CAN NOT give the BEST output. Please try using Utilitarianism.')
    else:
        k = max(min_happiness,key=min_happiness.get)

        total_happiness = 0
        for user in users:
            for key,value in user['happiness'].items():
                if key == k:
                    total_happiness += value       
        print(k)
        print(total_happiness)


def DesertBased_Proportionalism(users):
    diff_user = {}
    for user in users:
        for key,value in user['contribution'].items():
            # print(key)
            if key in diff_user.keys():
                diff_user[key].append(value - user['luck'])
            else:
                diff_user[key] = [value - user['luck']]
    
    print(diff_user)
    maxvalue = {}
    tmp = []
    for key,val in diff_user.items():
        maxvalue[key] = max(val)
        for n in val:
            tmp.append(n)
    # print(tmp)
    if tmp.count(max(maxvalue.values()))>1 :
        print('This Principle CAN NOT give the BEST output. Please try another.')
    else:
        print(max(maxvalue,key=maxvalue.get))
        print(max(maxvalue.values()))


def Libertarian_Proportionalism(users):
    max_contribution = defaultdict(lambda: float('-inf'))
    for user in users:
        for key,value in user['contribution'].items():
            max_contribution[key] = max(max_contribution[key], value)
    
    print(max_contribution)
    tmp = []
    for val in max_contribution.values():
        tmp.append(val)
    if tmp.count(max(max_contribution.values()))>1 :
        print('This Principle CAN NOT give the BEST output. Please try another.')
    else:
        print(max(max_contribution,key=max_contribution.get))
        print(max(max_contribution.values()))



def Egalitarianism(diction,principle):
    std = {}
    tmp = []
    for key,value in diction.items():
        std[key] = np.std(value)
        tmp.append(np.std(value))
    print(std)

    if tmp.count(min(std.values())) > 1:
        new_distri = {}
        tmp2 = []
        if principle == 1:
            for key,value in diction.items():
                if np.std(value) == min(std.values()):
                    new_distri[key] = max(value)
                    tmp2.append(max(value))
            if tmp2.count(max(new_distri.values())) > 1:
                print('This Principle CAN NOT give the BEST output. Please try another.')
            else:
                print(max(new_distri,key=new_distri.get))
                print(max(new_distri.values()))
        elif principle == 2:
            for key,value in diction.items():
                if np.std(value) == min(std.values()):
                    new_distri[key] = min(value)
                    tmp2.append(min(value))
            if tmp2.count(min(new_distri.values())) > 1:
                print('This Principle CAN NOT give the BEST output. Please try another.')
            else:
                print(min(new_distri,key=new_distri.get))
                print(min(new_distri.values()))
    else:
        print(min(std,key=std.get))
        print(min(std.values()))


def Autonomy_Egalitarianism(users):
    distri = {}
    for user in users:
        for key,value in user['autonomy'].items():
            if key in distri.keys():
                distri[key].append(value)
            else:
                distri[key] = [value]
    print(distri)
    Egalitarianism(distri,principle=1)


def NonMaleficence_Egalitarianism(users):
    distri = {}
    for user in users:
        for key,value in user['harm'].items():
            if key in distri.keys():
                distri[key].append(value)
            else:
                distri[key] = [value]
    print(distri)
    Egalitarianism(distri,principle=2)

def Equality_Opportunity(users):
    distri = {}
    for user in users:
        for key,value in user['opportunity'].items():
            if key in distri.keys():
                distri[key].append(value)
            else:
                distri[key] = [value]
    print(distri)
    Egalitarianism(distri,principle=1)


```
